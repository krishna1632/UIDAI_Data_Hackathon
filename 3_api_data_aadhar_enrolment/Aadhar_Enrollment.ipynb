## Aadhar Enrollment Data


```python
import pandas as pd
```

```python
df1 = pd.read_csv("api_data_aadhar_enrolment_0_500000.csv")
df2 = pd.read_csv("api_data_aadhar_enrolment_500000_1000000.csv")
df3 = pd.read_csv("api_data_aadhar_enrolment_1000000_1006029.csv")
```

```python
df1["date"] = pd.to_datetime(df1["date"], format="%d-%m-%Y")
df2["date"] = pd.to_datetime(df2["date"], format="%d-%m-%Y")
df3["date"] = pd.to_datetime(df3["date"], format="%d-%m-%Y")
```

```python
df1.sort_values(by="date")
```

```python
df2.sort_values(by="date")
```

```python
df3.sort_values(by="date")
```

### Conclusions:

- Enrollment update refers to enrollment of new citizens
- The dataset contains new enrollment records from 2nd March 2025 to 31st December 2025.
- The data records enrollments at state, district, and pincode level for each date.
- Enrollment data is of people of age group less than 5, from 5-17 and age group of 18+
- There are some 0 values in `age_0_5` `,age_5_17` and `age_18_greater` which indicate that no enrollment was performed for that particular age group on that date at that location.
