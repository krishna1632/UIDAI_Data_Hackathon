#### Conclusion

- We have data starting from 1st March 2025 to 29th December 2025
- Biometric update records are available across states and districts for different dates, with some locations showing zero update activity on certain days.
- One column is of age group of 5 years - 17 years
- Another column is of age group of 17 years+
- There are some 0 values in `bio_age_5_17` and `bio_age_17_` which indicates no biometric updation is done on that particular date
